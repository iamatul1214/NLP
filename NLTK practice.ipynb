```python
## Now store that stop word result into the another list
def remove_stop_words(list_of_tokens):
    from nltk.corpus import stopwords
    stop_words_en = set(stopwords.words('english'))
    refined_tokens = [words for words in list_of_tokens if words not in stop_words_en]
    return refined_tokens
stop_Words_Removed_tokens = remove_stop_words(token_list)
print(stop_Words_Removed_tokens)
```

```python
def remove_punctuation(list_of_tokens):
    from string import punctuation
    punctuation_set = set(punctuation)
    filtered_token = [words for words in list_of_tokens if words not in punctuation_set]
    return filtered_token
punctuation_removed_tokens = remove_punctuation(stop_Words_Removed_tokens)
punctuation_removed_tokens
```

```python
 ## stemming 
from nltk.stem import PorterStemmer
porter = PorterStemmer()

for word in punctuation_removed_tokens:
    print(porter.stem(word))
```

```python
stemming = [porter.stem(word) for word in punctuation_removed_tokens]
stemming
```

```python
## lemmitization
from nltk.stem import WordNetLemmatizer

lem = WordNetLemmatizer()

for word in punctuation_removed_tokens:
    print(lem.lemmatize(word))
```

```python
lemmitization = [lem.lemmatize(word) for word in punctuation_removed_tokens]
lemmitization
```

```python
## creating a dataframe which will store tokenized summary, stop words removed summary, punctuation removed summary, stemmed summary and lemmitized summary
text_mined_summary = pd.DataFrame()
text_mined_summary
```
