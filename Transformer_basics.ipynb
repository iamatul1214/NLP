```python
classifier_zero_shot = pipeline("zero-shot-classification")
## this model kind of unsupervised text classification analysis, where we can provide our own labels and it will classify on those labels only.
```

```python
classifier_zero_shot(
    "I love to visit pubs on Friday nights rather than saturday nights, because saturday nights are over crowded",
    candidate_labels=["Happy_sentiment","sad_sentiment"]
)

## the output here will show two probabilities the the two given labels
```

```python
classifier_zero_shot(
    "I prefer to visit pubs on Friday nights rather than saturday nights, because saturday nights are over crowded",
    candidate_labels=["Happy_sentiment","sad_sentiment"]
)

## We just used the word prefer rather than love and we can see the probability of happy sentiment has decreased from 56 to 51
```

```python
classifier_zero_shot(
    "I have to visit pubs on Friday nights rather than saturday nights, because saturday nights are over crowded",
    candidate_labels=["Happy_sentiment","sad_sentiment"]
)

## We just used the word have rather than prefer and we can see the probability of happy sentiment has decreased from 51 to 42 and sad sentiment has won
```

## These models have biasness because the corpus of data on which they are trained were bias, we can have a small demo for how to counter this, we can use masking and the output of masking can be used for classifier_zero_shot

```python
unmasking("This women works as a <mask>")
```

```python
generated_text = generator("This black man is a")[0]["generated_text"]
generated_text
```

```python
classifier_zero_shot(
generated_text,
    candidate_labels=["biased", "unbiased", "racist"],
)
```

```python
def predict_from_masking(sentence):
  try:

    predicted_values = unmasking(sentence)[0].items()
    for k,v in predicted_values:
      sentence=v
    return v
  except Exception as e:
    print("Error occured while predicting from masking",e)
```

```python
for k,v in predicted_values:
  if k == 'sequence':
    sentence = v
print(sentence)
```

```python
def classifying_from_zero_shot(sentence):
  try:
    print(classifier_zero_shot(
        sentence,
        candidate_labels=["unbiased", "biased","racist","sexist","misogynist"],
    ))
  except Exception as e:
    print("Error occured while classifying from zero shot", e)
```

```python
sentence_to_predict = "Coffee is <mask> better than tea"
sentence = predict_from_masking(sentence=sentence_to_predict)
classifying_from_zero_shot(sentence=sentence)
```
